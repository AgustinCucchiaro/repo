```python
import numpy as np
import math as m
```

```python
def OptionsVal(n, S, K, r, q, v, T, PC):
    dt = T/n                    
    u = m.exp(v*m.sqrt(dt)) 
    d = 1/u                     
    p = (m.exp((r-q)*dt)-d)/(u-d)   
    Pm = np.zeros((n+1, n+1))   
    Cm = np.zeros((n+1, n+1))
    tmp = np.zeros((2,n+1))
    for j in range(n+1):
        tmp[0,j] = S*m.pow(d,j)
        tmp[1,j] = S*m.pow(u,j)
    tot = np.unique(tmp)
    c = n
    for i in range(c+1):
        for j in range(c+1):
            Pm[i,j-c-1] = tot[(n-i)+j]
        c=c-1
    for j in range(n+1, 0, -1):
        for i in range(j):
            if (PC == 1):                               
                if(j == n+1):
                    Cm[i,j-1] = max(K-Pm[i,j-1], 0)     
                else:
                    Cm[i,j-1] = m.exp(-.05*dt) * (p*Cm[i,j] + (1-p)*Cm[i+1,j]) 
            if (PC == 0):                               
                if (j == n + 1):
                    Cm[i,j-1] = max(Pm[i,j-1]-K, 0)     
                else:
                    Cm[i,j-1] = m.exp(-.05*dt) * (p*Cm[i,j] + (1-p)*Cm[i+1,j])  
    return [Pm,Cm]



S = 30
k = 30
r = .06
q = 0
v = .28
T = 3
n = 12
PC = 0
OptionsVal(n,S,k,r,q,v,T,PC)
Pm,CmC = OptionsVal(n, S, k, r, q, v, T, PC)
PC = 0
_,CmP= OptionsVal(n, S, k, r, q, v, T, PC)
print('Pricing:\n',np.matrix(Pm.astype(float)))
print('Call Option:\n',np.matrix(CmC.astype(float)))
print('Put Option:\n',np.matrix(CmP.astype(float)))
```
